```python
import os
import ffm
import pandas as pd
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
```

```python
def read_dataset(filename):
    dataset = pd.read_csv(filename)
    X=[]
    y=[]
    country_set = []
    os_platform_set = []
    os_family_set = []
    session_depth_set = []
    for arr in dataset.itertuples():
        #print arr
        if arr[1] not in country_set:
            country_set.append(arr[1])

        if arr[2] not in os_platform_set:
            os_platform_set.append(arr[2])

        if arr[3] not in os_family_set:
            os_family_set.append(arr[3])

        if arr[4] not in session_depth_set:
            session_depth_set.append(arr[4])


    for arr in dataset.itertuples():
        #print arr
        x_list = []
        
        #country_code
        field = 1
        index = country_set.index(arr[1])
        value = 1

        x_list.append((field, index, value))
        
        #os_name
        
        field = 2
        index = os_platform_set.index(arr[2])
        value = 1

        x_list.append((field, index, value))

        #os_family
        '''
        field = 3
        index = os_family_set.index(arr[3])
        value = 1
        
        x_list.append((field, index, value))
        '''
        #session_depth
        field = 4
        index = session_depth_set.index(arr[4])
        value = 1
        
        x_list.append((field, index, value))

        X.append(x_list)
        y.append(arr[5])

    return X, y
```

```python
def ffm_trainer(train_X, train_y):
    print('Training processing!')


    ffm_data = ffm.FFMData(train_X, train_y)


    # train the model for 10 iterations

    n_iter = 100

    model = ffm.FFM(eta=0.1, lam=0.0001, k=10)
    model.init_model(ffm_data)

    for i in range(n_iter):
        print('iteration %d, ' % i)
        model.iteration(ffm_data)

        y_pred = model.predict(ffm_data)
        auc = roc_auc_score(train_y, y_pred)
        print('train auc %.4f' % auc)
        # save the model 
        
    model.save_model('ololo.bin')

    # load it to reuse the model
    model = ffm.read_model('ololo.bin')
    return model
```

```python
def ffm_prediction(model, test_X, test_y):
    print('\nTest validation!')

    
    ffm_test_data = ffm.FFMData(test_X, test_y)
    y_pred = model.predict(ffm_test_data)
    auc = roc_auc_score(test_y, y_pred)
    print('train auc %.4f' % auc)

```

```python
# prepare the data
# (field, index, value) format

X = [[(1, 2, 1), (2, 3, 1), (3, 5, 1)],
     [(1, 0, 1), (2, 3, 1), (3, 7, 1)],
     [(1, 1, 1), (2, 3, 1), (3, 7, 1), (3, 9, 1)]]

y = [1, 1, 0]

model = ffm_trainer(X, y)
ffm_prediction(model, X, y)
```
